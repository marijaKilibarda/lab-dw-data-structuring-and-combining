```python
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
df_marketing = pd.read_csv(url)
df_marketing
```

```python
# Step 3: Identify and Handle Missing Values
print("\nCount of null values in each column:")
df_marketing.isnull().sum()

# the strategy for handling missing values:
# fill missing numerical values with the mean and categorical values with the mode.
for column in df_marketing.columns:
    if df_marketing[column].isnull().any():  # if there are null values
        if df_marketing[column].dtype == 'object':  # categorical variable
            mode_value = df_marketing[column].mode()[0]  # get the mode
            df_marketing[column].fillna(mode_value, inplace=True)  # fill with mode
        else:  # Numerical variable
            mean_value = df_marketing[column].mean()  # get the mean
            df_marketing[column].fillna(mean_value, inplace=True)  # fill with mean


print("\nCount of null values in each column after handling:")
df_marketing.isnull().sum()
```

```python
null_counts = df_02.isnull().sum()
print(null_counts[null_counts > 0])
```

```python
#Step 4: Format Data Types
# Unnamed:_0: Already int64, no change needed.
# {customer, state, education, employmentstatus, location_code, marital_status, policy, renew_offer_type, 
#sales_channel, vehicle_class, vehicle_size:} 
#These are object types and might be better as category if they have a limited number of unique values.
# {customer_lifetime_value, income, monthly_premium_auto, months_since_last_claim, months_since_policy_inception, 
#number_of_open_complaints, number_of_policies, total_claim_amount, month:} 
#Already int64, no change needed unless you need them as floats for specific calculations.
# {response, gender, policy_type, vehicle_type}: Already category, no change needed.
# effective_to_date: Convert to datetime for date operations.

category_columns = ['customer', 'state', 'education', 'employmentstatus', 'location_code', 
                    'marital_status', 'policy', 'renew_offer_type', 'sales_channel', 
                    'vehicle_class', 'vehicle_size']

for col in category_columns:
   df_marketing[col] = df_marketing[col].astype('category')

numeric_cols = df_marketing.select_dtypes(include=['float64']).columns
df_marketing[numeric_cols] = df_marketing[numeric_cols].fillna(0).astype(int)  # fill NaN with 0 before conversion

# converting effective_to_date to datetime
df_marketing['effective_to_date'] = pd.to_datetime(df_marketing['effective_to_date'])
df_marketing.dtypes
```

```python
# Creating the pivot table
pivot_table = df_marketing.pivot_table(
    index='sales_channel',               # Rows
    values='total_claim_amount',         # Values to aggregate
    aggfunc='sum',                       # Aggregation function
    margins=False,                       # No total row
    observed=False        
).round(2)   

print("\nTotal Revenue by Sales Channel:")
pivot_table
```

```python
#Step 4: Create a Pivot Table for Average Customer Lifetime Value by Gender and Education Level

#Next, we will create a pivot table to show the average customer lifetime value per gender and education level. 

average_clv_by_gender_education = df_marketing.pivot_table(
    index='gender',                      # Rows
    columns='education',                 # Columns
    values='customer_lifetime_value',    # Values to aggregate
    aggfunc='mean',                      # Aggregation function
    fill_value=0,                        # Fill missing values with 0
    observed=False                        
).round(2)                              


# Display the pivot table
print("\nAverage Customer Lifetime Value by Gender and Education Level:")
print(average_clv_by_gender_education)
```
